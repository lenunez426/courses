```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics

#source: https://www.geeksforgeeks.org/how-to-plot-normal-distribution-over-histogram-in-python/
    
```

```python
##### PERIOD 1 #####

data_steps1 = np.array([370,310,376,369,343,360,399,370,436,349,399,305,262,337,332,315])
  
# Fit a normal distribution to
# the data:
# mean and standard deviation
mu, std = norm.fit(data_steps1) 
  
# Plot the histogram.
plt.hist(data_steps1, bins=25, density=True)
  
# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
  
plt.plot(x, p, 'k', linewidth=2)
title = "Steps Fit Values: {:.2f} and {:.2f}".format(mu, std)
plt.title(title)
  
plt.show()


data_steplength1 = np.array([0.18,0.5,0.31,0.58,0.18,0.42,0.45,0.3,0.4,0.4,0.54,0.51,0.6,0.51,0.51,0.45])

# Fit a normal distribution to
# the data:
# mean and standard deviation
mu, std = norm.fit(data_steplength1) 
  
# Plot the histogram.
plt.hist(data_steplength1, bins=25, density=True)
  
# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
  
plt.plot(x, p, 'k', linewidth=2)
title = "Step Length Fit Values: {:.2f} and {:.2f}".format(mu, std)
plt.title(title)
  
plt.show()
```
